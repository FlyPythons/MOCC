```python
def _plot_annotate(ax, v, color_dict):
    
    from matplotlib.colors import ListedColormap
    import matplotlib.patches as mpatches
    import numpy as np
    
    c = []
    v_dict = {}
    
    n = 1
    grades = []
    max_n = 0
    for i in v:
        s = []
        for j in str(i).split(";"):
            
            if j not in v_dict:
                v_dict[j] = n
                if j not in color_dict:
                    print(j)
                    break
                c.append(color_dict[j])
                
                n += 1
            s.append(v_dict[j])
        grades.append(s)
        max_n = max(max_n, len(s))
    
    y = np.zeros((max_n, len(grades)))
    
    for m, i in enumerate(grades):
        
        i = i + [i[-1]]*(max_n-len(i))
  
        for n, j in enumerate(i):
            y[n][m] = j
            
    c = ListedColormap(c)
    
    ax.imshow(y[::-1], aspect='auto',cmap=c)
    handers = []
    for k in color_dict:
        handers.append( mpatches.Patch(color=color_dict[k], label=k))
    #ax.legend(handles=handers,loc='upper left', borderaxespad=0., bbox_to_anchor=(1.05, 1),)
    #ax.legend()
            
    pass


def plot_meta(meta, names, color_dict):
    
    nrows = len(names)
    figh = nrows*0.5
    fig, axs = plt.subplots(nrows=nrows , figsize=(8, figh))
    
    fig.subplots_adjust(top=0.9, bottom=0.1,
                       left=0.2, right=0.99)

    for ax, name in zip(axs, names):
        _plot_annotate(ax, meta[name], color_dict[name])
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)
    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axs:
        ax.set_axis_off()
```

```python
def cal_rna2dna_distace(dna_df, rna_df, distance):
    r = 0
    n = 0
    for i in rna_df.index:
        _sample = rna_df.loc[i, "Sample"]
        _chr = rna_df.loc[i, "Human chromosome integrated"]
        _type = rna_df.loc[i, "Integration type"]
        _loc = rna_df.loc[i, "Human integration location"]
        _exp = []
        for j in dna_df[(dna_df.Sample == _sample) & (dna_df["Human chromosome integrated"] == _chr) & (dna_df["Integration type"] == _type)].index:
            if abs(_loc-dna_df.loc[j, "Human integration location"]) <= distance:
                r += 1
                break
        n += 1
    
    return r*1.0/n
```

```python
x = range(0, 100000, 1000)
y = []
for i in x:
    y.append(cal_rna2dna_distace(dna_df, rna_df, i))
```

```python
plt.subplots(figsize=(5.1,5))
plt.plot([i/1000 for i in x], [i*100 for i in y], "-",c="black")
plt.xlabel("Distance (kb)")
plt.ylabel("Percent (%)")
```
